```python
import pandas as pd
import numpy as np
import io
```

```python
# 1. Read the raw text from the CSV
with open('data.csv', 'r') as file:
    raw_text = file.read()

# 2. Strip out all the problematic double quotes
clean_text = raw_text.replace('"', '')

# 3. Load it into a pandas DataFrame
df = pd.read_csv(io.StringIO(clean_text))
```

```python
df.head()
```

```python
df.columns
```

```python
window = 3
df['rolling_mean'] = df['close'].rolling(window=window).mean()
```

```python
df.head()
```

```python
rolling_mean_avg = df['rolling_mean'].fillna(0.0)
rolling_mean_avg
```

```python
signal = np.where(
    df['close'] > rolling_mean_avg, 1, 0
)
```

```python
signal.mean()
```
